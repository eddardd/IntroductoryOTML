```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'custom' 
plt.rcParams['mathtext.rm'] = 'Bitstream Vera Sans' 
plt.rcParams['mathtext.it'] = 'Bitstream Vera Sans:italic' 
plt.rcParams['mathtext.bf'] = 'Bitstream Vera Sans:bold' 
plt.rcParams['font.size'] = 16 
plt.rcParams['mathtext.fontset'] = 'stix' 
plt.rcParams['font.family'] = 'STIXGeneral' 
```

```python
X = np.random.randn(30, 2)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].scatter(X[:, 0], X[:, 1])
axes[0].set_xlim([-3, 3])
axes[0].set_ylim([-3, 3])
axes[0].set_title('Available Data')
axes[0].axis('off')

sns.kdeplot(X[:, 0], X[:, 1], ax=axes[1], shade=True)
axes[1].set_title('Underlying Distribution')
axes[1].axis('off')

plt.savefig('../Figures/DataVsDistr.svg')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax = sns.kdeplot(X[:, 0], X[:, 1], shade=True, ax=ax)
ax.text(0.5, 0.5, s=r'$E$', horizontalalignment='center')
ax.text(1.5, 3, s="High Probability Region", horizontalalignment='center')
ax.arrow(1.5, 3, dx=(0.5 - 1.5) * 0.9, dy=(0.5 - 3) * 0.9)
ax.axis('off')

plt.savefig('../Figures/Pdist.svg')
```

```python
import mpl_toolkits.mplot3d.art3d as art3d
fig = plt.figure()

Z = np.array([1 / len(X)] * len(X))
ax = fig.add_subplot(111, projection='3d')

for xi, yi, zi in zip(X[:, 0], X[:, 1], Z):        
    line=art3d.Line3D(*zip((xi, yi, 0), (xi, yi, zi)), marker='o', markevery=(1, 1))
    ax.add_line(line)

ax.set_xlim3d(-3, 3)
ax.set_ylim3d(-3, 3)
ax.set_zlim3d(0, 0.05)

plt.savefig('../Figures/EmpiricalMeasure.svg')
```
